## Q1

```python
#    1. Most expensive trip (total amount)


q = """
        SELECT MAX(Total_amount) as 'Max Total Amount' 
        FROM nyc_data_sep
        
        
    """

pysqldf(q)
```

```python
#    1. Most expensive trip (total amount)


q = """
        SELECT * FROM nyc_data_sep
        ORDER BY Total_amount DESC
        LIMIT 1
    """

pysqldf(q)
```

## Q2

```python
#    2. Most expensive trip per mile (total amout/mile)

q = """
        SELECT Total_amount/Trip_distance AS amount_per_mile
        FROM nyc_data_sep
        ORDER BY amount_per_mile desc
        LIMIT 1
        
        
    """

pysqldf(q)
```

# Q4

```python
# Longest trip duration

q = """
        SELECT CAST(STRFTIME('%s',Lpep_dropoff_datetime) AS duration)-CAST(STRFTIME('%s',lpep_pickup_datetime) AS duration) AS Differance
        FROM nyc_data_sep
        ORDER BY Differance DESC
        LIMIT 1
        
    """

pysqldf(q)
```

```python
q = """
        SELECT *,CAST(STRFTIME('%s',Lpep_dropoff_datetime) AS duration)-CAST(STRFTIME('%s',lpep_pickup_datetime) AS duration) AS Differance
        FROM nyc_data_sep
        ORDER BY Differance DESC
        LIMIT 1
        
    """

pysqldf(q)
```

```python
q = """
        SELECT *,MAX(CAST(STRFTIME('%s',Lpep_dropoff_datetime) AS duration)-CAST(STRFTIME('%s',lpep_pickup_datetime) AS duration)) AS 'Differance'
        FROM nyc_data_sep
        WHERE 'Differance' = (SELECT MAX('Differance') FROM nyc_data_sep)
        
    """

pysqldf(q)
```

# Q5

```python
# mean tip by hour

q =  """
        SELECT lpep_pickup_hour AS Hours, avg(Tip_amount) 
        FROM nyc_data_sep
        GROUP BY Hours
    """

pysqldf(q)
```

```python
# mean tip by hour

q =  """
        SELECT strftime('%H', lpep_pickup_datetime) AS Hours, avg(Tip_amount) 
        FROM nyc_data_sep
        GROUP BY Hours
    """

pysqldf(q)
```

```python
# Median trip cost 

# q =  """
        
#         SELECT Total_amount
#         FROM nyc_data_sep
        
#     """
# pysqldf(q)


def median(my_list):
    n = len(my_list)
    my_list.sort()
    if n%2 ==0:
        med_p = my_list[n//2]
        med_n = my_list[n//2-1]
        med = (med_p+med_n)/2
    else:
        med=my_list[n//2]

    return med

```

# Q7

```python
# Average total trip by day of the week 

q =  """
        SELECT day, AVG(no_of_trip) FROM(
        SELECT strftime('%Y%m%d', lpep_pickup_datetime) as 'date', lpep_pickup_day_of_week as day,
                  COUNT(*) as 'no_of_trip'
        FROM nyc_data_sep
        GROUP BY date)
        GROUP BY day

         
    """

pysqldf(q)
```

# Q8

```python
# count of trips by hour
q =  """
        
        SELECT lpep_pickup_hour AS 'hours', COUNT(*) as 'no_of_trip'
        FROM nyc_data_sep
        GROUP BY hours

         
    """

pysqldf(q)
```
